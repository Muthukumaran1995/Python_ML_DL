### flight_delays.csv

```python
flight_data =pd.read_csv('flight_delays.csv', index_col='Month')
```

```python
flight_data[:3]
```

```python
plt.figure(figsize=(16, 8))
plt.title('Average arrival delays for the airline WN, by month')
#bar graph
sns.barplot(x=flight_data.index, y=flight_data['WN'])
#add the albels
plt.xlabel("Months")
plt.ylabel("Arrival delays")
```

### Heatmap 

```python
plt.figure(figsize=(10, 8))
plt.title('Average arrival delays for the airline WN, by month')
#bar graph
sns.heatmap(data=flight_data, annot=True)
#add the albels
plt.xlabel("Months")
plt.ylabel("Arrival delays")
```

```python
flight_data.corr()
```

```python
plt.figure(figsize=(10, 8))
plt.title('Average arrival delays for the airline WN, by month')
#bar graph
corr=flight_data.corr()
sns.heatmap(data=flight_data, annot=True, cmap='PRGn')
#add the albels
plt.xlabel("Months")
plt.ylabel("Arrival delays")
```
